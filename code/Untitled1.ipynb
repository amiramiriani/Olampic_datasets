```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/athletes_details.csv')
sport_df = pd.read_csv('../data/sports_scored.csv')
```

```python
df
```

```python
df = df.dropna(subset=['Sex', 'Age' , 'Height', 'Weight'])
```

```python
name = df['Sport'].unique()
print(name)
num_sports = np.size(name)
print(num_sports)
```

```python
sport_df_rank = sport_df[['Sport', 'Rank']].rename(columns={'Rank': 'Sport_Rank'})
merged_df = pd.merge(df,sport_df_rank , on='Sport', how='inner')
merged_df
```

```python
import pandas as pd

# Assume 'Year' is the Olympic year and 'Medal' is current/previous medal info
# We'll take the latest year as the "current Olympics"
current_year = merged_df['Year'].max()

# Flag athletes who won medals in previous years
merged_df['Has_Previous_Medal'] = merged_df.apply(
    lambda row: merged_df[
        (merged_df['Athlete'] == row['Athlete']) &
        (merged_df['Year'] < current_year) &
        (merged_df['Medal'].notnull())
    ].shape[0] > 0,
    axis=1
)

merged_df.to_csv('../data/athletes_sport_details.csv', index=False)
merged_df
```
