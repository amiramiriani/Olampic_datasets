```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/athletes_details.csv')
sport_df = pd.read_csv('../data/sports_scored.csv')
```

```python
df
```

```python
df = df.dropna(subset=['Sex', 'Age' , 'Height', 'Weight'])
```

```python
name = df['Sport'].unique()
print(name)
num_sports = np.size(name)
print(num_sports)
```

```python
sport_df_rank = sport_df[['Sport', 'Rank']].rename(columns={'Rank': 'Sport_Rank'})
merged_df = pd.merge(df,sport_df_rank , on='Sport', how='inner')
merged_df
```

```python
first_medal_year = {}

# Initialize the column
merged_df['Has_Previous_Medal'] = False

# Iterate through rows
for i, row in merged_df.iterrows():
    name = row['Name']
    year = row['Year']
    
    # Check if athlete has won a medal before this year
    if name in first_medal_year and year > first_medal_year[name]:
        merged_df.at[i, 'Has_Previous_Medal'] = True
    
    # Update the first medal year if this row has a medal
    if pd.notnull(row['Medal']):
        if name not in first_medal_year:
            first_medal_year[name] = year

# Save to CSV
merged_df.to_csv('../data/athletes_sport_details.csv', index=False)

merged_df
```
